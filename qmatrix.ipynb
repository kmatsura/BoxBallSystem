```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import re as rexprs
import sympy
from sympy import *
from sympy import init_printing
init_printing()
```

```python
def Pi(m, q, z):
    return m, q**m*z**m

def Qi(m, q, z):
    return m+1, z**m

def Ri(m, q, z):
    return m-1, (1 - q**(2*m))*z**m

def calc_amplitude(p1,p2, simple = True, symbol = "z", takefirstfomula = True): #p1,p2 is int or str or np.array
    if type(p1) == int:
        p1 = np.array(list(str(p1)),dtype = int)
    if type(p2) == int:
        p2 = np.array(list(str(p2)),dtype = int)

    if type(p1) == str:
        p1 = np.array(list(p1),dtype = int)
    if type(p2) == str:
        p2 = np.array(list(p2),dtype = int)        

        if len(p1) != len(p2):
            print("the length between p1 and p2 is not same")
            return 0
    
    m = sympy.Symbol("m")
    q = sympy.Symbol("q")
    z = 1
    #m = 0
    #z = sympy.Symbol(symbol)
    ans = 1
    for i in range(len(p1)):
        if p1[i] == 0:
            if p2[i] == 0:
                m, ans_temp = Pi(m,q,z)
            elif p2[i] == 1:
                m, ans_temp = Ri(m,q,z)
            else:
                print("error")
                break
        elif p1[i] == 1:
            if p2[i] == 0:
                m, ans_temp = Qi(m,q,z)
            elif p2[i] == 1:
                m, ans_temp_temp = Pi(m,q,z)
                ans_temp = ans_temp_temp * (-q)
            else:
                print("error")
                break
        else:
            print("error")
            break
        
        ans *= ans_temp
    #answer = sympy.summation(ans, (m, 0, sympy.oo))
    answer = ans
    answer = sympy.simplify(answer)
    if takefirstfomula == True:
        answer = take_firstfomula(answer)
    return answer

def take_firstfomula(x):
    xs = str(x)
    result = rexprs.search(r'\(\(', xs) 
    if result == None: return x
    start = result.span()[1]
    result = rexprs.search(r', And', xs)
    if result == None:
        result = rexprs.search(r',', xs)
    if result == None:
        result = rexprs.search(r'&', xs)
    if result == None: return x
    end = result.span()[0]
    struse = xs[start:end]
    #from IPython.core.debugger import Pdb; Pdb().set_trace()
    return sympy.sympify(struse)
```
