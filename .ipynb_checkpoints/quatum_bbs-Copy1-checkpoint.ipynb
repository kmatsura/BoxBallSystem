## calculate amplitude

```python
def P(m, q, z):
    return m, q**m*z**m

def Q(m, q, z):
    return m+1, z**m

def R(m, q, z):
    return m-1, (1 - q**(2*m))*z**m


def calc_amplitude(p1,p2, simple = True, symbol = "z"): #p1,p2 is int or str or np.array
    if type(p1) == int:
        p1 = np.array(list(str(p1)),dtype = int)
    if type(p2) == int:
        p2 = np.array(list(str(p2)),dtype = int)

    if type(p1) == str:
        p1 = np.array(list(p1),dtype = int)
    if type(p2) == str:
        p2 = np.array(list(p2),dtype = int)        

        if len(p1) != len(p2):
            print("the length between p1 and p2 is not same")
            return 0
    
    m = sympy.Symbol("m")
    q = sympy.Symbol("q")
    z = sympy.Symbol(symbol)
    ans = 1
    for i in range(len(p1)):
        if p1[i] == 0:
            if p2[i] == 0:
                m, ans_temp = P(m,q,z)
            elif p2[i] == 1:
                m, ans_temp = R(m,q,z)
            else:
                print("error")
                break
        elif p1[i] == 1:
            if p2[i] == 0:
                m, ans_temp = Q(m,q,z)
            elif p2[i] == 1:
                m, ans_temp_temp = P(m,q,z)
                ans_temp = ans_temp_temp * (-q)
            else:
                print("error")
                break
        else:
            print("error")
            break
        
        ans *= ans_temp
    answer = sympy.summation(ans, (m, 0, sympy.oo))
    answer = sympy.simplify(answer)
    return answer        


def calc_amplitude2(p1,p2, one, simple = True, verbose = False): #p1,p2 is int or str or np.array
    
    def P(m, q, z):
        if verbose == True:
            print("0to0")
        return m, q**m*z**m

    def Q(m, q, z):
        if num_one == m:
            if verbose == True:
                print("trap")
            return m, 0
        else:
            if verbose == True:
                print("+1")
            return m+1, z**m

    def R(m, q, z):
        if m == 0:
            return 0, 0
        else:
            if verbose == True:
                print("-1")
            return m-1, (1 - q**(2*m))*z**m

    if type(p1) == int:
        p1 = np.array(list(str(p1)),dtype = int)
    if type(p2) == int:
        p2 = np.array(list(str(p2)),dtype = int)

    if type(p1) == str:
        p1 = np.array(list(p1),dtype = int)
    if type(p2) == str:
        p2 = np.array(list(p2),dtype = int)        

        if len(p1) != len(p2):
            print("the length between p1 and p2 is not same")
            return 0
    
    num_one = np.sum(p1)
    q = sympy.Symbol("q")
    #z = sympy.Symbol("z")
    z=1
    answer = 0
    
    for m in range(one + 1):
        if verbose == True:
            print("m=" + str(m))
        ans = 1
        for i in range(len(p1)):
            if p1[i] == 0:
                if p2[i] == 0:
                    m, ans_temp = P(m,q,z)
                elif p2[i] == 1:
                    m, ans_temp = R(m,q,z)
                else:
                    print("error")
                    break
            elif p1[i] == 1:
                if p2[i] == 0:
                    m, ans_temp = Q(m,q,z)
                elif p2[i] == 1:
                    m, ans_temp_temp = P(m,q,z)
                    ans_temp = ans_temp_temp * (-q)
                else:
                    print("error")
                    break
            else:
                print("error")
                break
            
            ans *= ans_temp
        answer += ans
        if verbose == True:
            print(answer)
    answer = sympy.simplify(answer)
    return answer        
```

```python
path = "./data/result1-20.tex"
pathlist = ["1000","0100","0010","0001"]

import datetime
def writeexchange2result(filepath,pathlist):
    Ty = []
    for path1 in pathlist:
        temp = []
        for path2 in pathlist:
            temp.append(calc_amplitude(path1, path2, symbol = "y"))
        Ty.append(temp)

    Tz = []
    for path1 in pathlist:
        temp = []
        for path2 in pathlist:
            temp.append(calc_amplitude(path1, path2, symbol = "z"))
        Tz.append(temp)
        
    def multi(A,B):
        M = []
        for m in range(len(A)):
            templist = []
            for n in range(len(A)):
                temp = 0
                for i in range(len(A)):
                    temp += A[i][n] * B[m][i]
                templist.append(temp)
            M.append(templist)
        return M
    
    def sub(A,B):
        S = []
        for i in range(A):
            templist = []
            for j in range(A):
                templist.append(A[i][j] - B[i][j])
            S.append(templist)
        return S
    S = sub(multi(Ty, Tz), multi(Tz, Ty)) 
    dt_now = datetime.datetime.now()
    with open(path, mode = "a") as f:
        f.write("-"*50)
        f.write("\\\\")
        f.write("\n")
        f.write(str(dt_now))
        f.write("\\\\")
        f.write("\n")
        f.write("pathlist = ")
        f.write(str(pathlist))
        f.write("\\\\")
        f.write("\n")
        f.write("$$")
        f.write("substraction is ")
        f.write(sympy.latex(S))
        f.write("$$")
        f.write("\n")
        f.write("-"*50)
        f.write("\\\\")
        f.write("\n")
```

```python
path = "./data/result1-20.tex"
pathlist = ["1000","0100","0010","0001"]

writeexchange2result(path,pathlist)
```
